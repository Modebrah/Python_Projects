```python
# Height unit and data type conversion
def clean_height(height):
    try:
        # Extract numerical values from the string
        height_value = float(''.join(c for c in height if c.isdigit() or c == '.'))

        # Check if 'cm' is present in the string
        if 'cm' in height:
            return int(height_value)
        # Convert inches to cm (1 inch = 2.54 cm)
        elif 'in' in height:
            return int(height_value * 2.54)
        else:
            return height_value
    except ValueError:
        # Handle the case where height cannot be converted to a number
        return height

# Apply the function to the 'Height' column
fifa['Height'] = fifa['Height'].apply(clean_height).astype(int)
```

```python
# Weight unit and data type conversion
def heavy_weight(weight):
    try:
        # Extract numerical values from the string
        weight_value = float(''.join(c for c in weight if c.isdigit() or c == '.'))

        # check if 'kg' is present in string
        if 'kg' in weight:
            return int(weight_value)
        # convert lbs to kg (1lbs = 0.453592 kg)
        elif 'lbs' in weight:
            return int(weight_value * 0.45392)
        else:
            return weight_value
    except ValueError:
        # Handle the case where weight cannot be converted to a number
        return weight

# Apply the function to the 'Weight' column
fifa['Weight'] = fifa['Weight'].apply(heavy_weight).astype(int)
```

```python
#converting value column and replacing symbols/letter
def convert(value):
    if isinstance(value, float):
        return value  # Return the float value as is
    if 'M' in value:
        return float(value.replace('€', '').replace('M', '')) * 1000000
    elif 'K' in value:
        return float(value.replace('€', '').replace('K', '')) * 1000
    else:
        return None

# Apply the functions to the 'Wage', 'Release Clause', and 'Value' columns
fifa['Wage'] = fifa['Wage'].apply(convert)
fifa['Release Clause'] = fifa['Release Clause'].apply(convert)
fifa['Value'] = fifa['Value'].apply(convert)
```

```python
#Average Wage by Club
average_wage_by_club = fifa.groupby('Club')['Wage(€)'].mean().reset_index()
top_10_clubs = average_wage_by_club.sort_values(by='Wage(€)', ascending=False).head(10)

plt.figure(figsize=(12,6))
ax = sns.barplot(data=top_10_clubs, x='Club', y='Wage(€)', palette='Blues_d')
# Annotate the bars with their values
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:,.0f}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.title('Average Wage by Club')
ax.set_yticks([])
ax.set_yticklabels([])
plt.xticks(rotation=45)
plt.show()
```

```python
#Players with High Value and Low Wages
high_value = 1e7
low_wage = 50000
fifa['HighValue_LowWage'] = ((fifa['Value(€)'] > high_value) & (fifa['Wage(€)'] < low_wage))
plt.figure(figsize=(8,6))
ax = sns.countplot(data=fifa, x='HighValue_LowWage', palette='pastel')
# Annotate the bars with their values
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:,.0f}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.title('Number of Players with High Value and Low Wages')
plt.xlabel('High Value and Low Wages')
plt.ylabel('Count')
ax.set_yticks([])
ax.set_yticklabels([])
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

# CONCLUSION

In the culmination of our comprehensive analysis and visualization of the FIFA dataset, several key insights have emerged, shedding light on various aspects of player demographics, club dynamics, and performance-related trends. This analytical endeavor aimed to uncover patterns, correlations, and noteworthy findings within the dataset.

### Age Distribution:
The analysis revealed a predominant age distribution among players, with a significant concentration in the 20-25 age range. This observation suggests a strategic preference by national teams and clubs for younger players, possibly due to their perceived longevity and potential for sustained performance.

### Height Statistics:
Approximately 94% of players in the dataset exhibited a height within the range of 180cm to 182cm. This uniformity in height distribution may indicate a specific physical profile often sought after in the world of football.

### Club Affiliation:
Notably, 237 players are currently not affiliated with any club. Meanwhile, clubs from Europe's major leagues demonstrated a higher player representation compared to their counterparts in lesser-known leagues. This underscores the concentration of talent within the football powerhouses.

### Correlation between Total Stats and Wages:
A positive correlation was identified between a player's total statistics and their corresponding wages. This finding suggests that players with higher overall performance statistics tend to command higher wages, affirming the significance of on-field prowess in determining financial remuneration.

### Geographic Representation:
England emerged as the country with the highest representation of players in the dataset, closely followed by Germany. This geographical distribution underscores the global nature of football talent and the diversity of player origins.

### Foot Preference:
Examining player foot preferences revealed that 76.1% of players favor their right foot, while 23.9% opt for their left foot. This insight into player footedness provides valuable information for strategic considerations during matches.

### Player Positions:
Among the various player positions, the analysis identified CB (Center Back) as the most prevalent best position, with a notable concentration of players in this role. In contrast, fewer players identified CF (Center Forward) as their preferred position.

### Club Wage Bills:
Real Madrid emerged with the highest average wage bill, standing at €156,233.33, followed closely by FC Barcelona at €123,727.27. These figures illuminate the financial commitments of top-tier clubs in maintaining competitive squads.

### High Value, Low Wage Anomaly:
A noteworthy anomaly was observed in 3.3% of players, comprising 630 individuals, who possess high market values but receive comparatively low wages. This intriguing finding prompts further investigation into the factors influencing the valuation and compensation of these players.

In summary, this analysis has not only provided a nuanced understanding of the demographic and performance dynamics within the FIFA dataset but also uncovered intriguing anomalies that warrant deeper exploration. The insights gained contribute valuable perspectives for stakeholders in the football ecosystem, including clubs, scouts, and enthusiasts, guiding strategic decisions and further research endeavors in the dynamic world of football analytics.
