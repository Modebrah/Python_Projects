```python
def clean_height(height):
    try:
        # Extract numerical values from the string
        height_value = float(''.join(c for c in height if c.isdigit() or c == '.'))

        # Check if 'cm' is present in the string
        if 'cm' in height:
            return int(height_value)
        # Convert inches to cm (1 inch = 2.54 cm)
        elif 'in' in height:
            return int(height_value * 2.54)
        else:
            return height_value
    except ValueError:
        # Handle the case where height cannot be converted to a number
        return height

# Apply the function to the 'Height' column
fifa['Height'] = fifa['Height'].apply(clean_height).astype(int)
```

```python
def heavy_weight(weight):
    try:
        # Extract numerical values from the string
        weight_value = float(''.join(c for c in weight if c.isdigit() or c == '.'))

        # check if 'kg' is present in string
        if 'kg' in weight:
            return int(weight_value)
        # convert lbs to kg (1lbs = 0.453592 kg)
        elif 'lbs' in weight:
            return int(weight_value * 0.45392)
        else:
            return weight_value
    except ValueError:
        # Handle the case where weight cannot be converted to a number
        return weight

# Apply the function to the 'Weight' column
fifa['Weight'] = fifa['Weight'].apply(heavy_weight).astype(int)
```

```python
def convert(value):
    if isinstance(value, float):
        return value  # Return the float value as is
    if 'M' in value:
        return float(value.replace('€', '').replace('M', '')) * 1000000
    elif 'K' in value:
        return float(value.replace('€', '').replace('K', '')) * 1000
    else:
        return None

# Apply the functions to the 'Wage', 'Release Clause', and 'Value' columns
fifa['Wage'] = fifa['Wage'].apply(convert)
fifa['Release Clause'] = fifa['Release Clause'].apply(convert)
fifa['Value'] = fifa['Value'].apply(convert)
```
